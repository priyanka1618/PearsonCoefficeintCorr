```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Read the data which is downloaded form kaggle and put it into a dataframe 
SuicideRate = pd.read_csv('C:\\Users\\sharmpri\\Downloads\\data Science\\machines learning\\Corr_coefiecient\\master.csv\\master.csv')
print(SuicideRate)
```

```python
#read the headers with some records from the dataframe using head function
print(SuicideRate.head())
```

```python
# Calculating pearson coefficient 
pearsonCorr = SuicideRate.corr(method = 'pearson')
pearsonCorr # paiwise correlation beween all the columns
```

```python
# it is more time consuming to interpret if data is more
import seaborn as sb # seaborn is the python library for visalizaing data . it is on the top of matplotlib library

# to look and making the data easier to interpret add the pearson coefficient of correlation in the heatmap
#heatmap--> plot rectangular color-encoded matrix 
sb.heatmap(pearsonCorr, # data
            xticklabels = pearsonCorr.columns, # plotting the columns names of the data
            yticklabels = pearsonCorr.columns, #plotting the columns names of the data
            cmap = 'RdBu_r', # list of colors used in the heatmap
            linewidth = 0.5, # width of the line devide the each cell
             annot = True
             )
```
